```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

# **1. Tiền xử lý và chuẩn bị dữ liệu:**

## *1.1.  Tổng quan về dữ liệu*

```python
df = pd.read_excel("Global Superstore.xls")
```

```python
# Lọc và trích xuất dữ liệu theo quốc gia Germany
data = df[df['Country'] == 'Germany'].copy()

# Hiển thị dữ liệu
data
```

```python
# xem thông tin chung về dữ liệu
data.info()
```

## *1.2. Bỏ biến dư thừa*

```python
# Xóa cột 'Postal Code'
data.drop('Postal Code', axis=1, inplace=True)
data.info()
```

```python
data.iloc[:,4:].describe()
```

## *1.3. Xử lý trùng lắp và dữ liệu thiếu*

### **1.3.1. Kiểm tra trùng lắp**

```python
# đếm số lượng trùng lắp
data.duplicated().sum()
```

### **1.3.2. Kiểm tra dữ liệu thiếu**

```python
#thay thế '?', 'Unknowns' bằng NaN
data.replace('?', np.nan, inplace=True)
data.replace('unknown', np.nan, inplace=True)
```

```python
# Kiểm tra missing value
data.isnull().sum()
```

## *1.4. Chuẩn hóa dữ liệu*

```python
# Chuẩn hoá giá trị chưa phù hợp thành định dạng phù hợp (ngày tháng, tiền tệ…)
# Chuẩn hóa ngày tháng
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

# Chuyển đổi kiểu dữ liệu sang số thực (float)
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data['Discount'] = pd.to_numeric(data['Discount'], errors='coerce')
data['Shipping Cost'] = pd.to_numeric(data['Shipping Cost'], errors='coerce')

# Chuyển đổi kiểu dữ liệu sang số nguyên (int)
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce').astype(int)

# Hiển thị thông tin về dữ liệu sau khi chuẩn hóa
data.head()
```

## *1.5. Xử lý ngoại lai*

*Nhóm nhận thấy việc các thành phố lớn có dân số đông hơn, mức độ tiêu thụ nhiều hơn các tỉnh thành phố nhỏ rất nhiều. Điều này khiến dữ liệu sẽ có sự cách biệt lớn, nhưng các dữ liệu này không phải là dữ liệu ngoại lai. Nếu xử lý ngoại lai sẽ làm dữ liệu sai về mặt thực tế. Do đó nhóm quyết định không xử lý dữ liệu ngoại lai ạ!*

## *1.6. Thêm biến phái sinh*

```python
# 1. Thêm biến 'Profit Margin' (Tỷ suất lợi nhuận)
data['Profit Margin'] = (data['Profit'] / data['Sales']).replace([np.inf, -np.inf], 0).fillna(0)

# 2. Thêm biến 'Delivery Speed' (Thời gian giao hàng)
data['Delivery Speed'] = (data['Ship Date'] - data['Order Date']).dt.days

# 3. Thêm biến 'Sales per Customer' (Doanh số trung bình mỗi khách hàng)
data['Sales per Customer'] = data.groupby('Customer ID')['Sales'].transform('mean')

# Hiển thị thông tin sau khi thêm biến
data.head()
```

# **2. Mô tả dữ liệu:**

## *2.1. Tổng quan thị trường Đức*

```python
# Tổng quan thị trường Đức
total_orders = data['Order ID'].nunique()
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
total_customers = data['Customer ID'].nunique()
average_Profit_Margin = data['Profit Margin'].mean()
average_Delivery_Speed = data['Delivery Speed'].mean()
average_Sales_per_Customer = data['Sales per Customer'].mean()

print("Tổng quan thị trường Đức:")
print(f"Tổng số đơn hàng: {total_orders}")
print(f"Tổng doanh số: {total_sales:.2f}")
print(f"Tổng lợi nhuận: {total_profit:.2f}")
print(f"Tổng số khách hàng: {total_customers}")
print(f"Tỷ suất lợi nhuận trung bình: {average_Profit_Margin:.2f}")
print(f"Thời gian giao hàng trung bình: {average_Delivery_Speed:.2f} ngày")
print(f"Doanh số trung bình mỗi khách hàng: {average_Sales_per_Customer:.2f}")
```

## *2.2. Top 5 thành phố có doanh thu cao nhất*

```python
# Doanh thu thành phố theo bang
city_state_sales = data.groupby(['State', 'City'])['Sales'].sum().reset_index()
city_state_sales = city_state_sales.rename(columns={'Sales': 'Total Sales'})

# Top 5 thành phố có doanh thu cao nhất
top_5_cities = city_state_sales.sort_values(by='Total Sales', ascending=False).head(5)
print("Top 5 thành phố có doanh thu cao nhất:")
top_5_cities
```

## *2.3. Top 5 thành phố có lượng khách cao nhất*

```python
# Lượng khách hàng ở mỗi thành phố theo bang
city_state_customers = data.groupby(['State', 'City'])['Customer ID'].nunique().reset_index()
city_state_customers = city_state_customers.rename(columns={'Customer ID': 'Total Customers'})

# Top 5 thành phố có nhiều khách hàng nhất
top_5_cities = city_state_customers.sort_values(by='Total Customers', ascending=False).head(5)
print("Top 5 thành phố có nhiều khách hàng nhất:")
top_5_cities
```

## *2.4. Top 5 thành phố có lợi nhuận cao nhất*

```python
# Lợi nhuận ở mỗi thành phố theo bang
city_state_profit = data.groupby(['State', 'City'])['Profit'].sum().reset_index()
city_state_profit = city_state_profit.rename(columns={'Profit': 'Total Profit'})

# Top 5 thành phố có lợi nhuận cao nhất
top_5_cities = city_state_profit.sort_values(by='Total Profit', ascending=False).head(5)
print("Top 5 thành phố có lợi nhuận cao nhất:")
top_5_cities
```

## *2.5. Thống kê danh mục sản phẩm*

```python
# Phân tích thống kê danh mục sản phẩm
category_stats = data.groupby('Category').agg(
    total_sales=('Sales', 'sum'),
    total_profit=('Profit', 'sum'),
    total_quantity=('Quantity', 'sum'),
    average_profit_margin=('Profit Margin', 'mean')
)
print("\n Thống kê danh mục theo doanh thu, lợi nhuận, số lượng và lợi nhuận biên")
category_stats = category_stats.rename(
    columns={
        'total_sales': 'Total Sales',
        'total_profit': 'Total Profit',
        'total_quantity': 'Total Quantity',
        'average_profit_margin': 'Average Profit Margin'
    }
)

# Hiển thị kết quả
category_stats
```

## *2.6. Thống kê danh mục phụ*

```python
# Phân tích thống kê danh mục phụ (Sub Category)
sub_category_stats = data.groupby('Sub-Category').agg(
    total_sales = ('Sales', 'sum'),
    total_profit = ('Profit', 'sum'),
    total_quantity = ('Quantity', 'sum'),
    average_profit_margin = ('Profit Margin', 'mean')
)
print("\nThống kê loại mặt hàng theo doanh thu, lợi nhuận, số lượng và lợi nhuận biên")
sub_category_stats = sub_category_stats.rename(
    columns={
        'total_sales': 'Total Sales',
        'total_profit': 'Total Profit',
        'total_quantity': 'Total Quantity',
        'average_profit_margin': 'Average Profit Margin'
    }
)

# Hiển thị kết quả
sub_category_stats
```

## *2.7. Thống kê phân khúc khách hàng*

```python
# Phân tích thống kê phân khúc khách hàng
segment_stats = data.groupby('Segment').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Customers=('Customer ID', 'nunique'),
    Avg_Sales_per_Customer=('Sales per Customer', 'mean')
)
print("\nThống kê theo phân khúc khách hàng:")
segment_stats = segment_stats.rename(
    columns={
        'Total_Sales': 'Total Sales',
        'Total_Profit': 'Total Profit',
        'Total_Customers': 'Total Customers',
        'Avg_Sales_per_Customer': 'Average Sales per Customer'
    }
)
# Hiển thị kết quả
segment_stats
```

# **3. Trực quan hóa dữ liệu:**

## *3.1. Xu hướng doanh thu và lợi nhuận theo thời gian*

```python
# Doanh thu theo thời gian
plt.figure(figsize=(12, 6))

# Calculate monthly and quarterly sales data
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Order Month'] = data['Order Date'].dt.to_period('M')
monthly_sales = data.groupby('Order Month')['Sales'].sum()

data['Order Quarter'] = data['Order Date'].dt.to_period('Q')
quarterly_sales = data.groupby('Order Quarter')['Sales'].sum()

plt.figure(figsize=(12, 6))

# Vẽ doanh số hàng tháng
plt.plot(mdates.date2num(monthly_sales.index.to_timestamp().to_pydatetime()), monthly_sales.values, label='Doanh số hàng tháng')

# Vẽ doanh số hàng quý, giữ nguyên trục x để trùng với doanh số hàng tháng
plt.plot(mdates.date2num(quarterly_sales.index.to_timestamp().to_pydatetime()), quarterly_sales.values, label='Doanh số hàng quý', marker='o', linestyle='--', color='orange')

plt.xlabel('Tháng/Năm')
plt.ylabel('Doanh số')
plt.title('Xu hướng doanh thu theo thời gian')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Nhận xét:** Nhìn chung doanh số có xu hướng tăng cho thấy khả năng kinh doanh tốt, phát triển. Doanh số hàng tháng có nhiều biến biến động, theo xu hướng tăng nhưng chưa ổn định. Biểu hiện rõ nhất trong giai đoạn 7 tháng ( T5 - T11 năm 2014) biến động tăng, giảm đến 8 lần, trong khi giai đoạn 8 tháng (T1 - T8 năm 2011) lại biến động tăng. Doanh số hàng quý biến động nhẹ, theo xu hướng tăng ổn định hơn.

```python
# Chuyển đổi cột 'Order Date' sang kiểu datetime
data.loc[:, 'Order Date'] = pd.to_datetime(data['Order Date'])

# Tạo cột 'Order Month' từ 'Order Date'
data.loc[:, 'Order Month'] = data['Order Date'].dt.to_period('M')

# Tính toán lợi nhuận hàng tháng
monthly_profit = data.groupby('Order Month')['Profit'].sum()

# Tạo cột 'Order Quarter' từ 'Order Date'
data.loc[:, 'Order Quarter'] = data['Order Date'].dt.to_period('Q')

# Tính toán lợi nhuận hàng quý
quarterly_profit = data.groupby('Order Quarter')['Profit'].sum()

# Vẽ lợi nhuận hàng tháng
plt.figure(figsize=(12, 6))
plt.plot(mdates.date2num(monthly_profit.index.to_timestamp().to_pydatetime()), monthly_profit.values, label='Lợi nhuận hàng tháng')

# Vẽ lợi nhuận hàng quý, giữ nguyên trục x để trùng với lợi nhuận hàng tháng
plt.plot(mdates.date2num(quarterly_profit.index.to_timestamp().to_pydatetime()), quarterly_profit.values, label='Lợi nhuận hàng quý', marker='o', linestyle='--', color='orange')

plt.xlabel('Tháng/Năm')
plt.ylabel('Lợi nhuận')
plt.title('Xu hướng lợi nhuận theo thời gian')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Nhận xét:** Tổng quan lợi nhuận có xu hướng tăng cho thấy khả năng vận hành, hoạt động tốt.
Lợi nhuận hàng tháng có xu hướng tăng dần theo thời gian, nhưng vẫn có nhiều biến động.
Lợi nhuận hàng quý cũng biến động theo xu hướng tăng,  một số thời điểm lợi nhuận tăng đột biến (khoảng đầu năm 2012 và giữa năm 2014), và cũng có một số thời điểm lợi nhuận giảm mạnh (giai đoạn cuối năm 2011 và giữa năm 2013).

→ Xu hướng doanh thu và lợi nhuận nhìn chung có điểm tương đồng nhau theo thời gian, nhưng rõ rệt nhất từ cuối năm 2023 đến gần cuối năm 2024 lợi nhuận có chiều giảm mạnh hơn so với doanh thu.

## *3.10. Tỷ lệ khách hàng quay lại so với khách hàng mua lần đầu*

```python
# Phân tích tỷ lệ khách hàng mua một lần (One-time customers) so với khách hàng quay lại (Repeat customers).

# Tính số lượng khách hàng mua một lần và khách hàng quay lại
customer_orders = data.groupby('Customer ID')['Order ID'].count()
one_time_customers = customer_orders[customer_orders == 1].count()
repeat_customers = customer_orders[customer_orders > 1].count()

# Tạo biểu đồ tròn
labels = ['Khách hàng mua lần đầu', 'Khách hàng quay lại']
sizes = [one_time_customers, repeat_customers]
explode = (0.1, 0)
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.title('Tỷ lệ khách hàng quay lại so với khách hàng mua lần đầu')
plt.show()
```

- Khách hàng quay lại chiếm 73.7%, cho thấy doanh nghiệp có tỷ lệ giữ chân khách hàng khá tốt.
- Khách hàng chỉ mua một lần chiếm 26.3%, phản ánh một phần tỷ lệ rời bỏ khách hàng (Churn Rate).

→ Nguồn thu đa phần từ khách hàng quay lại, đây là phần đáng cân nhắc, tỷ lệ 73,7% là khá cao, cần tiếp tục cải thiện chất lượng để khách hàng ngày càng thân thiết với doanh nghiệp. ## *3.13. Mức độ ưu tiên theo phân khúc khách hàng*

```python
# Phân tích mức độ ưu tiên theo phân khúc khách hàng
# Tạo nhóm dữ liệu Order Priority theo Segment
segment_priority = data.groupby(['Segment', 'Order Priority'])['Order ID'].count().reset_index()
segment_priority = segment_priority.pivot(index='Segment', columns='Order Priority', values='Order ID')
segment_priority = segment_priority.div(segment_priority.sum(axis=1), axis=0)

# Định nghĩa màu
priority_colors_segment = {
    'Critical': '#FF6F61',  # Màu đỏ cam
    'High': '#D3D3D3',      # Màu xám nhạt
    'Medium': '#6495ED',    # Màu xanh dương
    'Low': '#FFA07A'        # Màu cam nhạt
}

# Vẽ biểu đồ tròn Priority theo Segment
fig1, axes1 = plt.subplots(1, len(segment_priority.index), figsize=(16, 5))

for i, segment in enumerate(segment_priority.index):
    ax = axes1[i]
    colors = [priority_colors_segment[level] for level in segment_priority.columns]
    segment_priority.loc[segment].plot(kind='pie', autopct='%1.1f%%', ax=ax, legend=False, colors=colors)
    ax.set_title(f'Order Priority - {segment}')

plt.tight_layout()
plt.show()
```

**Nhận xét:**
- Cả ba phân khúc khách hàng có các mức độ ưu tiên gần như tương tự nhau, không có sự chênh lệch lớn. Mức độ Critical cao nhất là 8,4% của Home Office, mức độ High cao nhất là 28,4% của Consumer. 
- Home Office cũng là phân khúc có mức độ Medium và Low cao nhất, cho thấy phân khúc này đa dạng về nhu cầu ưu tiên.


## *3.14. Mức độ ưu tiên theo danh mục sản phẩm*

```python
# Phân tích mức độ ưu tiên theo danh mục sản phẩm
# Tạo nhóm dữ liệu Order Priority theo Category
category_priority = data.groupby(['Category', 'Order Priority'])['Order ID'].count().reset_index()
category_priority = category_priority.pivot(index='Category', columns='Order Priority', values='Order ID')
category_priority = category_priority.div(category_priority.sum(axis=1), axis=0)

# Định nghĩa màu mới
priority_colors_category = {
    'Critical': '#FFB6C1',  # Màu hồng nhạt
    'High': '#ADD8E6',      # Màu xanh nhạt
    'Medium': '#90EE90',    # Màu xanh lá nhạt
    'Low': '#FFDAB9'        # Màu cam nhạt
}
# Vẽ biểu đồ tròn Priority theo Category
fig2, axes2 = plt.subplots(1, len(category_priority.index), figsize=(16, 5))
for i, category in enumerate(category_priority.index):
    ax = axes2[i]
    colors = [priority_colors_category[level] for level in category_priority.columns]
    category_priority.loc[category].plot(kind='pie', autopct='%1.1f%%', ax=ax, legend=False, colors=colors)
    ax.set_title(f'Order Priority - {category}')
plt.tight_layout()
plt.show()
```

**Nhận xét:**
- Danh mục Furniture có mức độ Critical cao nhất, đạt 8,8%. Sở dĩ danh mục này có mức độ ưu tiên cao do nhu cầu sắm sửa lớn, hoặc thay thế cho các thiết bị trước bị hư hỏng nên khách hàng cần ưu tiên hàng đầu.
- Trong khi đó, mức độ High và Medium cao nhất thuộc về Office Supplies. Điều này khá dễ hiểu khi phân khúc Home Office có mức độ ưu tiên cao dẫn đến danh mục Office Supplies cũng theo đó mà tăng thêm.
